```python
import pandas as pd
```

```python
clientes    = pd.read_csv  (r"C:\Users\Note reserva\Desktop\Python\04. ETL Pandas\arquivos_csv\Clientes.csv",delimiter=",")
clientes.head(10)
```

```python
clientes.dtypes
```

```python
# Converter a coluna 'data' para o tipo de dados datetime, se ainda não estiver(para funcionar na query tem que estar nesse formato de data)
clientes['created_at'] = pd.to_datetime(clientes['created_at'])
```

```python
clientes.head()
```

```python
#clientes['ano']=clientes['created_at'].dt.year
```

```python
# filtrando 1 ano especifico 
clientes.query('created_at.dt.year  == 2017')
```

```python
mesdez = clientes.query('created_at.dt.month  == 10')
```

```python
#exemplo ano 2017 e mes 10
clientes.query('created_at.dt.year == 2017  &  created_at.dt.month  == 10 ')
```

```python
#exemplo ano 2017 ou mes 10 (simulando obedecendo ou 1 criterio ouo outro) fique atento com e ou ou 
clientes.query('created_at.dt.year == 2017  |  created_at.dt.month  == 10 ')
```

```python
ordens      = pd.read_excel(r"C:\Users\Note reserva\Desktop\Python\arquivos_excel\Ordens.xlsx")
ordens.head()
```

```python
# mes e ano atual
data_atual=datetime.now()
ano_atual = datetime.now().year  #traz o ano atual
mes_atual = datetime.now().month # traz o mes atual
```
